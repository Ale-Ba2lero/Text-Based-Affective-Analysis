# Text based emotion analysis

## Introduction

With this analysis, we want to investigate the correlations between gaze features and emotions using NLP techniques.

Since there is no unique way to represent emotions, we decided to perform this analysis by using the two main approaches of emotion depiction, which adhere to the two schools of thought on emotions: 
the discrete model and the continuous one.

Additionally, we approached the analysis in two different ways. In the first phase, we analyzed the emotion score values (both continuous and discrete) of the single words, to see if the words alone could provide meaningful information even when taken without context. In the second phase, we considered not single words, but entire sentences into account.

## Dataset

### Discrete emotions lexicon
The first dataset we used is the "NRC Emotion Intensity Lexicon", it provides the emotional score of 6000 English words for the eight discrete emotions identified by this lexicon, which are: anger, fear, joy, sadness, disgust, anticipation, trust, and surprise.

### Continuous emotions lexicon
The second dataset we used is the "NRC VAD Lexicon" which adheres to the continuous approach of emotion representations by using three indicators to describe the emotional state of each individual. Those three indicators are Valence, Arousal, and Dominance. This lexicon contains more than 20.000 entries and provides for each one the three VAD score values.

### Eye-tracker recordings
We used the "ZuCo 2.0" dataset to get the eye-tacker data that consists of the recordings of the eye features of 18 people during the reading task of 739 English sentences.

The recorded features provided by the dataset are the following:

Gaze duration (GD): The sum of all fixations on the current word in the first-pass reading before the eye moves out of the word; 

Total reading time (TRT): The sum of all fixation durations on the current word, including regressions; 

First fixation duration (FFD): The duration of the first fixation on the prevailing word; 

Single fixation duration (SFD): The duration of the first and only fixation on the current word; 

Go-past time (GPT): The sum of all fixations prior to progressing to the right of the current word, including regressions to previous words that originated from the current word;

Mean pupil size (MPS)

Number of fixations (nFix)


*due to technical problems with the dataset the SFD values could not have been retrieved

### Features selection and Multicollinearity

We can notice that some of the eye-tracker features are strictly bound to each other. To avoid multicollinearity between independent variables we need to carefully understand and select the most relevant features, trying to avoid the ones that brings the same informations.

In some cases these features may or may not have the same value depending on the reading performance. This could lead to multicollinearity witch we want to avoid.

For example, the Total Reading Time consists on the Gaze Duration plus all the other fixations on the word. This implies a strong bound between the two features. We can think about TRT = GD + regressions. 
The First Fixation Duration features is very similar to GD and therefore the bound is similar.
It's easy to infer that the number of fixations is strictly bound to features that implies multiple fixations on the same word as Total Reading Time and Go Past Time.

The only features that seems independent is the Mean Pupil Size since it refers to a completelly different unit.

To better understand the bound within those feature, we are going to perform a multicollinearity analysis. We choose to calculate te VIF (variance inflation factor) for the eye-tracker features, that will help us to find those correlations.

#### VIF calculation

As previously anticipated, we will perform two differents analysis, word wise and sentence wise. For this reason we will handle the data in two different ways. We will perform the VIF analysis for both the resuting dataset, to see if the data handling have influenced on the features correlations.

```python
# Importing the eye-tracker dataset
zuco_ds_words = pd.read_csv('./Lexicons/ZuCo_words_dataset.csv')

zuco_ds_words = zuco_ds_words.iloc[:,1:]
zuco_ds_words = pd.DataFrame(StandardScaler().fit_transform(zuco_ds_words), columns=zuco_ds_words.columns, index=zuco_ds_words.index)
zuco_ds_words = zuco_ds_words.dropna()

# the independent variables set 
X_w = zuco_ds_words

# VIF dataframe 
vif_data = pd.DataFrame() 
vif_data["feature"] = X_w.columns 
  
# calculating VIF for each feature 
vif_data["VIF"] = [variance_inflation_factor(X_w.values, i) for i in range(len(X_w.columns))] 
  
print(vif_data)
```

```python
zuco_ds_sentences = pd.read_csv(r'./Lexicons/Emotion_Sentences_Cross_Analysis.csv')
zuco_ds_sentences = zuco_ds_sentences.iloc[:,11:]
zuco_ds_sentences = pd.DataFrame(StandardScaler().fit_transform(zuco_ds_sentences), columns=zuco_ds_sentences.columns, index=zuco_ds_sentences.index)
zuco_ds_sentences = zuco_ds_sentences.dropna()

# the independent variables set 
X_s = zuco_ds_sentences

#pd.set_option('display.max_rows', X_s.shape[0]+1)
#print (X_s)

# VIF dataframe 
vif_data_s = pd.DataFrame() 
vif_data_s["feature"] = X_s.columns 
  
# calculating VIF for each feature 
vif_data_s["VIF"] = [variance_inflation_factor(X_s.values, i) for i in range(len(X_s.columns))] 
  
print(vif_data_s)
```

For what concerns the word-based analysis results, no significant signs of multicollinearity have been discovered apart between the "total reading time" and "number of fixations" features.

Conversely, the sentence-based VIF analysis shows that all the features are heavily bound together (exception made for the "mean pupil size" feature). 

Now we must choose which features to keep for our analysis.

For the word-based analysis, since the only two highly correlated variables are TRT and nFix, we have to discard one of them. We decided to discard TRT for two principal reasons: TRT VIF value is higher than the nFix one, which means that by removing the TRT feature, we will decrease the correlation between all the variables more significantly. The second reason is due to the interpretation of the features. Since TRT is much more similar to the other features (they all measure reading times in different ways) we think that keeping nFix would add more information to the analysis.

In the case of the sentence based analysis, the problem is different. All the features, with the only exception of MPS, are very strongly correlated. For this reason, we decided to keep only one feature in addition to MPS, since this is the only way to avoid multicollinearity. The feature we decided to keep is TRT since we think that it is one of the most comprehensible and significant features.

Now let's repeat the VIF analysis to check whether our choices have had an impact on the multicollinearity between the remaining variables.

We start with the word-based VIF analysis:

```python
# the independent variables set 
X_w = zuco_ds_words[['MPS', 'GD', 'FFD', 'GPT', 'nFix']]

# VIF dataframe 
vif_data = pd.DataFrame() 
vif_data["feature"] = X_w.columns 
  
# calculating VIF for each feature 
vif_data["VIF"] = [variance_inflation_factor(X_w.values, i) for i in range(len(X_w.columns))] 

print(vif_data)
```

We can notice now that the VIF value of all the variables is below the threshold of 5. This indicates that there are no signs of significant multicollinearity between the features.

Let's repeat the analysis for the sentence based case:

```python
# the independent variables set 
X_s = zuco_ds_sentences[['MPS', 'TRT']]

# VIF dataframe 
vif_data_s = pd.DataFrame() 
vif_data_s["feature"] = X_s.columns 
  
# calculating VIF for each feature 
vif_data_s["VIF"] = [variance_inflation_factor(X_s.values, i) for i in range(len(X_s.columns))] 
  
print(vif_data_s)
```

We have obtained good results also in this case. No signs of significant multicollinearity are shown by the VIF values.

## Word-based OLS analysis

Now that we selected the features, let's start with the word-based analysis where we will search for meaningful relationships between the gaze features (MPS, GD, FFD, GPT, and nFix) and the single words emotion score values taken from the two lexicons. 

We calculated the mean value of the gaze features for each word that appears in each lexicon.

We then used a multiple linear regression model to look up for a linear relationship between the gaze features (independent variables) and each emotion score value (dependant variable).

By applying this model to each "emotion score value" (both discrete and continuous), we could identify which of those gaze features are related (and at what level) to them. 

### W.B. Discrete emotions analysis

```python
zuco_ds = pd.read_csv('./Lexicons/ZuCo_words_dataset.csv')
zuco_ds.head()
```

We generated the ZuCo_words_dataset by calculating the mean value of the eye-tracker features for each word that appears in the dataset.

```python
zuco_ds.iloc[:,1:] = pd.DataFrame(StandardScaler().fit_transform(zuco_ds.iloc[:,1:].values), columns=zuco_ds.columns[1:])
word_ds = zuco_ds[['Word','GD', 'FFD', 'GPT', 'nFix', 'MPS']]
word_ds.head()
```

```python
sns.pairplot(word_ds)
```

```python
# Import the discrete emotion lexicon.
anger_lex = pd.read_csv('./Lexicons/NRC_Emotion_Intensity_Lexicon/NRC-Emotion-Intensity-anger-scores.csv')

# Data standardization.
anger_lex.iloc[:,1:] = pd.DataFrame(StandardScaler().fit_transform(anger_lex.iloc[:,1:].values), columns=anger_lex.columns[1:])
anger_lex.head()
```

Each of the discrete emotion datasets contains a list of words and the corresponding score value for the emotion.

```python
# Dataframe merge
anger_ds = pd.merge(word_ds, anger_lex, how ='inner', on =['Word'])
anger_ds.head()
```

```python
sns.pairplot(anger_ds, x_vars=['GD', 'FFD', 'GPT', 'nFix', 'MPS'], y_vars=['Score'],)
```

The intersection of the two datasets contains only 44 entries, which are the words contained in the sentences provided by the ZuCo dataset witch are also present in the NRC anger dataset. (The other emotions lexicons results in a similar number of matching words).

By looking at the plots we do not expect to find strong signs of linear correlation.

We then perform a multiple linear regression analysis (OLS) using the anger score values as the dependant variable and the scores of the eye-tracker features as the independent ones.

```python
X = anger_ds[['GD', 'FFD', 'GPT', 'nFix', 'MPS']]
y = anger_ds['Score']
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
model.summary()
```

We try to improve the result by applying backward elimination setting the significance level at 0.5%.

We proceed by removing the independent variable with the highest p-value (if p-value > 0.05) and then fit the model again with the remaining variables.

We repeat the process until no independent variable has a p-value higher than 0.05, or no variable is left.

We wrote a script to speed up the process. The script shows for each step which is the variables considered and their p-value.

```python
model = be.backwardEliminationMLR(X,y, True)
```

In this case, even after the backward elimination process, we have found no feature with a satisfying level of relationship with the anger emotion score.

However, we must point out that the low number of matching words may have heavily affected the results.

We now repeat this analysis for the remaining emotions.
