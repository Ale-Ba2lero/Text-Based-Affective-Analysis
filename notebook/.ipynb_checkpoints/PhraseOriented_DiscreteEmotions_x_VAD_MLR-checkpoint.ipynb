```python
import pandas as pd
```

## Importing the  dataset

```python
dataset = pd.read_csv('Emotion_Sentences_Cross_Analysis.csv')

X = dataset[['Anger','Anticipation','Disgust','Fear','Joy','Sadness','Surprise','Trust']]
                
y_arousal = dataset.iloc[:, 0].values
y_valence = dataset.iloc[:, 1].values
y_dominance = dataset.iloc[:, 2].values
```

## Models building

```python
import statsmodels.api as sm
```

### Discrete emotions - Arousal

```python
X = sm.add_constant(X)
model_a = sm.OLS(y_arousal, X).fit()
model_a.summary()
```

### Discrete emotions - Valence

```python
model_v = sm.OLS(y_valence, X).fit()
model_v.summary()
```

### Discrete emotions - Dominance

```python
model_d = sm.OLS(y_dominance, X).fit()
model_d.summary()
```
