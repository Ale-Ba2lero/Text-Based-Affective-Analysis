### Take data from CSV files

```python
zuco_cs = pd.read_csv(r'../Lexicons/ZuCo_words_dataset.csv')
sentences_cs = pd.read_csv(r'../Lexicons/Emotion_Sentences_Cross_Analysis.csv')
#zuco_cs = zuco_cs.drop(index=430)
```

## Multiple Linear Regression between VAD emotions and Discrete emotions in sentences

With these analysis we want to search a possible correlation between the values of the single discrete emotion in each sentence.

We are going to exclude the sentences where the discrete emotion we are analyzing as depentent variable is = 0.

### Analysis: Anger as dependent and Arousal, Valence, Dominance as indipendent

```python
vad_anger_cs = pd.DataFrame(sentences_cs, columns=['Arousal','Valence','Dominance','Anger'])
vad_anger_cs = vad_anger_cs[vad_anger_cs['Anger'] > 0]

x = vad_anger_cs[['Arousal','Valence','Dominance']]
y = vad_anger_cs['Anger']
x = sm.add_constant(x)

model = be.backWardEliminationMLR(x,y)
model.summary()
#model = sm.OLS(y, x).fit()
#model.summary()
```

We can see a strong correlation between the dependent variable Anger and the independent variables Valence(p value = 0.008)

SL = 0.05

### Analysis: Sadness as dependent and Arousal, Valence, Dominance as indipendent
