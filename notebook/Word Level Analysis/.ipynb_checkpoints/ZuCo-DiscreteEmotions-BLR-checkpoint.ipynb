```python
import arviz as az
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pymc3 as pm
from pymc3 import traceplot
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

print('Running on PyMC3 v{}'.format(pm.__version__))
%config InlineBackend.figure_format = 'retina'
az.style.use('arviz-darkgrid')
```

## Importing the dataset

```python
zuco_ds = pd.read_csv('../Lexicons/ZuCo_words_dataset.csv')
anger_lex = pd.read_csv('../Lexicons/NRC_Emotion_Intensity_Lexicon/NRC-Emotion-Intensity-anger-scores.csv')
anger_ds = pd.merge(zuco_ds, anger_lex, how ='inner', on =['Word'])
anger_ds = anger_ds.drop(columns='Word')
```

## Data scaling

```python
scaler = MinMaxScaler()
anger_ds = pd.DataFrame(scaler.fit_transform(anger_ds), columns=anger_ds.columns)
```

## Model building

```python
X = anger_ds.iloc[:,:-1]
y = anger_ds.iloc[:,-1]
data = dict(x=X, y=y)
```

```python
with pm.Model() as model:
    # specify glm and pass in data. The resulting linear model, its likelihood and
    # and all its parameters are automatically added to our model.
    pm.glm.GLM.from_formula('y ~ X.MPS + X.TRT + X.GD + X.FFD', data)
    trace = pm.sample(3000, cores=2) # draw 3000 posterior samples using NUTS sampling
```

```python
plt.figure(figsize=(7, 7))
traceplot(trace)
plt.tight_layout();
```

```python
az.summary(trace, round_to=2)
```
